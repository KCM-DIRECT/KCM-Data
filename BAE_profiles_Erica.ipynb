## Iterate through all files to find matching serial number and store filename

```python
from os import listdir

directory = './BAE_profiles/'
modules = []
files = []

for file in listdir(directory):
    filename = file
    if filename.endswith('.csv'):
        data = pd.read_csv(directory + filename, error_bad_lines=False)
        if len(data) > 42:
            number = data.iloc[42,0]
            #print(type(number),number)
            if type(number) == str:
                #print(number)
                serial = number[19:44]
                if serial == '039-1523-03P-402335-00710':
                    modules.append(serial)
                    files.append(filename)
                else:
                    continue
        else:
            continue
    else:
        continue
        
#modules
```
