```python
import csv
import os
from os import listdir
```

```python
#making list csv file names in directory

directory = 'Raw data/profiles 2017-2019/'
csv_list = []

for file in listdir(directory):
    if file.endswith('.csv'):
        csv_list.append(file)
```

```python
len(csv_list)
```

```python
#fetching date cell
filename = '!3J0018_ProfileData_20170920082828.csv'

with open('Raw data/profiles 2017-2019/' + filename) as file:
    reader = csv.reader(file)
    for row in reader:
        if 'Data retrieved' in row[0]:
                print(filename, row)
```

```python
#trying to fetch date cells from all csv in directory
list_of_dates = []

for filename in csv_list:
    with open('Raw data/profiles 2017-2019/' + filename) as file:
        reader = csv.reader(file)
        for row in reader:
            try:
                if 'Data retrieved' in row[0]:
                    print(filename, row[0])
                    list_of_dates.append(row[0])
            except:
                pass #some files have no data
```
