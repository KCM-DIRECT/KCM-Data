```python
import csv
import os
from os import listdir
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### Only looking at Bus 1 now
directory = 'bus_1'
csv_list = []

for file in listdir(directory):
    if file.endswith('.csv'):
        csv_list.append(file)
#print(csv_list)
```

```python
## if error reading the files, use this to get address of the files
cwd = os.getcwd()
files = os.listdir(cwd+'/'+ directory)
for name in files:
    print(name)
```

```python
# Code from Sara
date_list=[]
name_list=[]
final_list=[]
serial_list = []
listoflist=[]
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth', None)


cols = ['Filename', 'DateRetrieved','ASCII#']
for name in files:
    if name.endswith('.csv'):
        file = open(cwd + '/'+directory+'/' + name)
        reader = csv.reader(file)
        for row in reader:
            for element in row:
                if 'Data retrieved' in element:
                    element_date = element[16:]
                    element_date = element_date.replace('@','')
                    date_list.append(element_date)
                    name_list.append(name)
                if 'Mfg Data (ASCII):' in element:
                    ASCII_num = element[18:47]
        serial_list.append(ASCII_num)
        #seems like it is not reading the correct ASCII to the corresponding file
        listoflist.append(serial_list)

final_list = list(zip(name_list,date_list,listoflist))
#result = [x[1] for x in final_list]
#result.sort(key=lambda date:datetime.strptime(date, "%m/%d/%Y %H:%M:%S"))
files_dates = pd.DataFrame(final_list,columns=cols)
files_dates['DateRetrieved'] = pd.to_datetime(files_dates.DateRetrieved)
files_dates.sort_values('DateRetrieved', inplace=True)
files_dates.reset_index(drop = True, inplace=True)
files_dates.T
#final_list
```
