```python
import csv
import os
from os import listdir
```

```python
#making list of csv filenames in directory

directory = 'Raw data/profiles 2017-2019/'
csv_list = []

for file in listdir(directory):
    if file.endswith('.csv'):
        csv_list.append(file)
```

```python
#general substring search function that searches every cell in the csv

some_list=[]
substring = '' #type what you want to look for here

def search_substring_csv(substring):
    for filename in csv_list: #every csv in directory
        with open('Raw data/profiles 2017-2019/' + filename) as file:
            reader = csv.reader(file)
            for row in reader:
                try:
                    for element in row:
                        if substring in element: #since all rows turn into list, i check all elements of every list
                        some_list.append(element)
                except:
                    pass #some files have no data
```

```python
#fetch date cells from all csvs in directory

list_of_dates = []
substring = 'Data retrieved'

for filename in csv_list:
    with open('Raw data/profiles 2017-2019/' + filename) as file:
        reader = csv.reader(file)
        for row in reader:
            try:
                for element in row:
                    if substring in element:
                        print(filename, '|', element)
                        list_of_dates.append(element)
            except:
                pass #some files have no data
```
