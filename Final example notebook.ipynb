```python
import altair as alt
import csv
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import os
from os import listdir
import pandas as pd
import re

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('heat map.csv')
df2 = df.drop(columns=['Unnamed: 0'])
df2
```

```python
#create grouped dataframe with module changes as variable
df_heatmap = df2.melt(id_vars=['Bus', 'Module', 'Date'])
```

```python
#disable altair max rows
alt.data_transformers.disable_max_rows()

#plot heat map for each bus that changes colour each time a module is replaced

select_bus = alt.selection_single(
    name='Select', fields=['Bus'], init={'Bus': 1},
    bind=alt.binding_range(min=1, max=130, step=1)
)

chart = alt.Chart(df_heatmap).mark_rect(stroke='black').encode(
    x=alt.X('Date', title="Date", sort=None),
    y=alt.Y('Module', title="Module", sort=None),
    color=alt.Color('value', legend=None)
).add_selection(select_bus).transform_filter(select_bus)

chart
```
